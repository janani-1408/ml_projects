###Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
dataset=pd.read_csv("/content/data.csv")   #Breast Cancer dataset
```

```python
dataset.head(10)
```

```python
dataset.columns
```

```python
dataset.shape
```

```python
x=dataset['radius_mean'].values.reshape(-1,1)
y=dataset["diagnosis"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
plt.scatter(dataset['radius_mean'],y)
```

```python
y_pred=model.predict([[20]])
y_pred
```
